```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("D:/RAG-App/Data/rahul.txt",encoding = 'UTF-8')
documents =  loader.load()
text_splitter = CharacterTextSplitter(chunk_size=500,chunk_overlap=30)
docs = text_splitter.split_documents(documents)
docs
```

```python
embeddings = OllamaEmbeddings(model="gemma:2b")
```

```python
pip install faiss-cpu
```

```python
db = FAISS.from_documents(docs,embeddings)
db
```

```python
#querying

query = "Gandhi completed his elementary education at?"
docs = db.similarity_search(query=query)
docs

```

# As a retriever

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs[0].page_content
```

# Similarity search with score

## lower score is better

```python
docs_and_search = db.similarity_search_with_score(query)
docs_and_search
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_and_score = db.similarity_search_by_vector(embedding_vector)
docs_and_score
```

```python
# Saving and Loading
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index",embeddings, allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query)
docs
```
